```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu mẫu
models = ["KNN", "SVM", "Logistic Regression"]
accuracy_scores = [0.8181818181818182, 0.7922077922077922, 0.7850162866449512]  # Accuracy
f1_scores = [0.6956521739130435, 0.6190476190476191, 0.6021505376344086
]  # F1-score
colors = ['blue', 'red', 'green']  # Màu cho từng mô hình

x = np.arange(len(models))  # Vị trí trên trục x
width = 0.3  # Độ rộng cột

fig, ax = plt.subplots(figsize=(8,5))

# Vẽ cột
bars1 = ax.bar(x - width/2, accuracy_scores, width, label="Accuracy", color='blue')
bars2 = ax.bar(x + width/2, f1_scores, width, label="F1-Score", color='red')

# Xác định giá trị cao nhất của Accuracy và F1-score
max_acc = max(accuracy_scores)
max_f1 = max(f1_scores)

# Lấy index của giá trị cao nhất để xác định màu
max_acc_index = accuracy_scores.index(max_acc)
max_f1_index = f1_scores.index(max_f1)

# Vẽ đường nét đứt cho Accuracy (màu xanh)
ax.axhline(y=max_acc, linestyle="dashed", color='blue', linewidth=1.5)
ax.text(len(models) - 0.5, max_acc + 0.01, f"{max_acc:.2f}", color='blue', fontsize=12, fontweight='bold')

# Vẽ đường nét đứt cho F1-score (màu đỏ)
ax.axhline(y=max_f1, linestyle="dashed", color='red', linewidth=1.5)
ax.text(len(models) - 0.5, max_f1 - 0.03, f"{max_f1:.2f}", color='red', fontsize=12, fontweight='bold')

# Gắn nhãn và tiêu đề
ax.set_xlabel("Thuật toán")
ax.set_ylabel("Giá trị")
ax.set_title("So sánh Accuracy và F1-Score")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.ylim(0, 1)  # Giới hạn trục y từ 0 đến 1
plt.show()
```
